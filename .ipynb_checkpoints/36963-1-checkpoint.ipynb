```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal,fft
```

```python
from thinkdsp import Wave,decorate

from DPI import hl2adb as hl2a
from DPI import changeDriver
# 0=driver mode   0=hl2a,1=local, 2=exl50,3=east,4=hl2m
changeDriver(0)
```

## Mpol_14,外中平面靠下位置，1160 ms左右，40-30 kHz下扫频

```python
a=1
ts, y, U = hl2a(36963, 'Mpol_14', 1.15, 1.2)
freq = 1/(ts[1]-ts[0])
ys = signal.detrend(y)
wave = Wave(ys, ts/1e3, freq)
```

```python
f,t,s=signal.spectrogram(ys,fs=freq,window='hamming',nperseg=1024*2,noverlap=1500,nfft=1024*4,detrend='linear',mode='psd',return_onesided=False)
fig, ax = plt.subplots(figsize=(3.54*3, 2.26*2))
plt.pcolormesh(t+ts[0],fft.fftshift(f),10*np.log10(fft.fftshift(s, axes=0)))
plt.ylim(0,80)
plt.set_cmap(plt.cm.jet)
# plt.clim(-70,-50)
plt.colorbar()
```

```python
segment = wave.segment(start=1.15, duration=0.85)
fig,ax = plt.subplots(figsize=(3.54*2, 2.26*2))
segment.plot()
decorate(xlabel='Time (s)',ylabel='Amplitude')
```

```python
sp = segment.make_spectrogram(1024)
fig,ax = plt.subplots(figsize=(3.54*2, 2.26*2))
sp.plot()
plt.set_cmap(plt.cm.jet)
plt.ylim(0,80)
plt.clim(0,1000)
plt.colorbar()
decorate(xlabel='Time (s)', ylabel='Frequency (Hz)')
```

## CPS 38 GHz, 对应频率范围响应

```python
ts, y, U = hl2a(36963, 'doppler_31', 1.15, 1.2)
freq = 1/(ts[1]-ts[0])
ys = signal.detrend(y)
ts2, y2, U2 = hl2a(36963, 'doppler_32', 1.15, 1.2)

y=signal.detrend(y)
y2=signal.detrend(y2)

fig,ax = plt.subplots(figsize=(3.54*2, 2.26*2))
ax.plot(ts,y,'b-')
ax.plot(ts2,y2,'r-')
```

```python
wave1 = y+y2*1J
f,t,s=signal.spectrogram(wave1,fs=freq,window='hamming',nperseg=1024,noverlap=800,nfft=4096,detrend='linear',mode='psd',return_onesided=False)
fig, ax = plt.subplots(figsize=(3.54*3, 2.26*2))
plt.pcolormesh(t+ts[0],fft.fftshift(f),10*np.log10(fft.fftshift(s, axes=0)))
plt.ylim(-150,150)
plt.set_cmap(plt.cm.jet)
plt.clim(-70,-20)
plt.colorbar()
```

# 软x射线确定空间位置

## 11通道，芯部

```python
def sxspect(chl,ts,te):
    ts, y, U = hl2a(36963, chl, ts/1e3, te/1e3)
    freq = 1/(ts[1]-ts[0])
    ys = signal.detrend(y)
    fig1,ax1 = plt.subplots(figsize=(3.54*2, 2.26*2))
    ax1.plot(ts,ys,'b-')
    f,t,s=signal.spectrogram(ys,fs=freq,window='hamming',nperseg=128,noverlap=100,nfft=256,detrend='linear',mode='psd',return_onesided=False)
    fig2, ax2 = plt.subplots(figsize=(3.54*3, 2.26*2))
    plt.pcolormesh(t+ts[0],fft.fftshift(f),10*np.log10(fft.fftshift(s, axes=0)))
    plt.ylim(0,50)
    plt.set_cmap(plt.cm.jet)
    # plt.clim(-70,-50)
    plt.colorbar()
    return ax1,ax2
```

```python
sxspect('sx11',1150,1170)
```

```python
sxspect('sx20',1150,1200)
```

SX边缘通道隐约可见，但可能是20kHz长寿模倍频？

# ECE
MDR_1------MDR_60,频率范围:62 GHz-91.5 GHz(60 道,间隔 0.5 GHz)

```python
def ecespect(chl,ts,te):
    ts, y, U = hl2a(36963, chl, ts/1e3, te/1e3)
    freq = 1/(ts[1]-ts[0])
    ys = signal.detrend(y)
    # fig1,ax1 = plt.subplots(figsize=(3.54*2, 2.26*2))
    # ax1.plot(ts,ys,'b-')
    f,t,s=signal.spectrogram(ys,fs=freq,window='hamming',nperseg=512,noverlap=400,nfft=1024,detrend='linear',mode='psd',return_onesided=False)
    fig2, ax2 = plt.subplots(figsize=(3.54*3, 2.26*2))
    plt.pcolormesh(t+ts[0],fft.fftshift(f),10*np.log10(fft.fftshift(s, axes=0)))
    plt.ylim(0,100)
    plt.set_cmap(plt.cm.jet)
    plt.clim(-80,-20)
    plt.colorbar()
    return ax2
```

```python
ecespect('MDR_2',1150,1200)
```

类似sx，边缘通道疑似倍频

# BES

BES01-32

```python
def besspect(chl,ts,te):
    ts, y, U = hl2a(36963, chl, ts/1e3, te/1e3)
    freq = 1/(ts[1]-ts[0])
    ys = signal.detrend(y)
    # fig1,ax1 = plt.subplots(figsize=(3.54*2, 2.26*2))
    # ax1.plot(ts,ys,'b-')
    f,t,s=signal.spectrogram(ys,fs=freq,window='hamming',nperseg=2048,noverlap=1800,nfft=2560,detrend='linear',mode='psd',return_onesided=False)
    fig2, ax2 = plt.subplots(figsize=(3.54*3, 2.26*2))
    plt.pcolormesh(t+ts[0],fft.fftshift(f),10*np.log10(fft.fftshift(s, axes=0)))
    plt.ylim(0,50)
    plt.set_cmap(plt.cm.jet)
    plt.clim(-70,-20)
    plt.colorbar()
    return ax2
```
